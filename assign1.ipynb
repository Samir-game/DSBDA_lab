### Integration

```python
my_csv1 = pd.read_csv('/home/csl-4/Downloads/diabetes_dataset.csv')
my_csv2 = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'Doctor': ['Dr.A', 'Dr.B', 'Dr.C', 'Dr.D', 'Dr.E'],
    'Hospital': ['H1', 'H2', 'H3', 'H4', 'H5']
})
my_csv1['id'] = range(1, len(my_csv1)+1)
merged_df = pd.merge(my_csv1, my_csv2, on='id', how='left')
print(merged_df.tail(50))
```

```python

my_csv_indexed = my_csv1.set_index('id')
doctor_indexed = my_csv2.set_index('id')


joined_df = my_csv_indexed.join(doctor_indexed, how='left')
print(joined_df.head())
```

```python
concat_df = pd.concat([my_csv1, my_csv2], ignore_index=True, axis=0)
print(concat_df)
```

```python
concat_df = pd.concat([my_csv1, my_csv2], axis=1)
print(concat_df)
```

```python
concat_df = pd.concat([my_csv1, my_csv2], keys=['dataset1','dataset2'])
print(concat_df)
```

```python
melted_df = pd.melt(concat_df,id_vars=['id'],
                    value_vars=['Glucose', 'BloodPressure', 'BMI', 'Age'],
                    var_name='Measurement', value_name='Value')

print(melted_df.head())
```

```python
pivot_table_df = melted_df.pivot_table(
    index='id',
    columns='Measurement',
    values='Value',
    aggfunc='mean'  # you can use sum, max, etc.
)
print(pivot_table_df.head())
```
