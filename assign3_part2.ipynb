```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np


iris = load_iris()


df = pd.DataFrame(iris.data, columns=iris.feature_names)

df["species"] = pd.Categorical.from_codes(iris.target, iris.target_names)

df.head()
```

```python
mean_table = df.groupby("species" , observed=True).mean()

print("\nMEAN VALUES (Species-wise)\n")
mean_table.round(2)
```

```python
std_table = df.groupby("species", observed=True).std()

print("\nSTANDARD DEVIATION (Species-wise)\n")
std_table.round(2)
```

```python
percentile_table = (
    df.groupby("species", observed=True)
      .quantile([0.25, 0.50, 0.75])
      .rename(index={0.25: "Q1 (25%)", 0.50: "Q2 (50%)", 0.75: "Q3 (75%)"})
)

print("\nPERCENTILE VALUES (Species-wise)\n")
percentile_table.round(2)
```
